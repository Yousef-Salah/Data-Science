## Pivotting the data into a 2D matrix to visualize it in groups

```python
data.groupby(['Regionname','Type']).size().rename("count").reset_index().pivot(index='Regionname', columns='Type',values='count')
```

```python
stats=data.groupby(['Regionname','Type']).size().rename("count").reset_index().pivot(index='Regionname', columns='Type',values='count').fillna(0)
```

```python
stats
```

```python
stats.plot.bar()
```

```python
stats.plot.bar(stacked=True)
```

```python
stats.plot.bar(subplots=True,figsize=(10,10))
```
