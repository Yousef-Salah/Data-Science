```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize you Graph
plt.figure(figsize=(5,3), dpi=300)

# plt.plot(x, y, label='2x', color='gray', linewidth=3, linestyle='--', marker='.', markersize=10, markeredgecolor='blue')
plt.plot(x, y, 'b^--', label='2x') # using shorthand notation



## second Line
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--', label='projection')


plt.title('Our First Graph', fontdict={'fontname':'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.savefig('myFirstGraph.png', dpi=500)

plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]
bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.show()
```
